# 맨홀비율, 배수펌프비율, 불투수면비율 Merge

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 행정구역 코드

```python
Code_DF = pd.read_excel('Reflux_data/병합/PIH_merge/구군 행정코드.xlsx',index_col=False)
Code_DF.head()
```

## 배수펌프 비율

```python
Pump_DF = pd.read_excel('Reflux_data/병합/PIH_merge/배수펌프비율.xlsx',index_col=False)
Pump_DF.head()
```

```python
Pump_DF.drop(['ZONE','ZONE_AREA','PUMP_COUNT'],axis=1,inplace=True)
Pump_DF.head()
```

## 불투수면 비율

```python
IMP_DF = pd.read_excel('Reflux_data/병합/PIH_merge/불투수면비율.xlsx',index_col=False)
IMP_DF.head()
```

```python
IMP_DF.drop(['ZONE','IMP_SUR','ZONE_AREA'],axis=1,inplace=True)
IMP_DF.head()
```

## 맨홀 비율

```python
Manhole_DF = pd.read_excel('Reflux_data/병합/PIH_merge/맨홀비율.xlsx',index_col=False)
Manhole_DF.head()
```

```python
Manhole_DF.drop(['ZONE','MANHOLES_COUNT','ZONE_AREA'],axis=1,inplace=True)
Manhole_DF.head()
```

## 병합

#### Code_DF와 Pump_DF 병합

```python
Main_data = pd.merge(Code_DF,Pump_DF,on='ADM_CD') 
Main_data.head()
```

#### Main_DF와 IMP_DF 병합

```python
Main_data = pd.merge(Main_data,IMP_DF,on='ADM_CD') 
Main_data.head()
```

#### Main_DF와 Manhole_DF 병합

```python
Main_data = pd.merge(Main_data,Manhole_DF,on='ADM_CD') 
Main_data
```

## 데이터 저장

```python
Main_data.to_excel('PIH_merge.xlsx')
```
