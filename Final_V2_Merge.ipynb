```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 침수가 되어 있는 시간을 기준으로 DIST=1

```python
F_Final_V2_01 = pd.read_excel('Reflux_data/Final_V2/F_Final_V2_01.xlsx',index_col=False)
F_Final_V2_01.head()
```

```python
# Nan 값을 0으로 바꿈
F_Final_V2=F_Final_V2.fillna(0)
F_Final_V2
```

```python
F_Final_V2.to_excel("Reflux_data/Final_V2/F_Final_V2_test1.xlsx")
```

## 비침수 데이터 

```python
Unf_Final_V2_01 = pd.read_excel('Reflux_data/Final_V2/Unf_Final_V2_01.xlsx',index_col=False)
Unf_Final_V2_01.head()
```

## 행 병합

```python
Final_V2_test01=pd.concat([F_Final_V2_01,Unf_Final_V2_01])
Final_V2_test01.head()
```

```python
Final_V2_test01.to_excel("Reflux_data/Final_V2/Final_V2_test01.xlsx")
```

## 침수 되어 있는 데이터만 상관관계 분석

- DIST와 관계 
 * HOUR_RAINFALL :0.39
 * SOLPE_AVG : 0.13

```python
plt.figure(figsize=(10,10))
sns.heatmap(data = F_Final_V2_01.corr(), annot=True, 
fmt = '.2f', linewidths=.5, cmap='Blues')
plt.show()
```
