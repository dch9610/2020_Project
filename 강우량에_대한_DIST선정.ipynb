## 강수량이 있으면 DIST값이 1인 조건

```python
F_Final_DF = pd.read_excel('Reflux_data/Final_V1/F_Final_test1.xlsx',index_col=False)
F_Final_DF.head()
```

```python
# 조건에 따른 열값 바꾸기
# - 시간당 강수량이 0이상이면 DIST 값이 1

F_Final_DF.loc[(F_Final_DF['HOUR_RAINFALL'] > 0),'DIST'] = 1
```

```python
# 조건에 따른 열값 바꾸기
# - 시간당 강수량이 0이면 DIST 값이 0

F_Final_DF.loc[(F_Final_DF['HOUR_RAINFALL'] == 0),'DIST'] = 0
```

```python
F_Final_DF.to_excel("Reflux_data/Final_V1/F_Final_V2_test1.xlsx")
```

## 다른 조건

```python
F_Final_DF = pd.read_excel('Reflux_data/Final_V1/F_Final_test1.xlsx',index_col=False)
F_Final_DF.head()
```

```python
# 조건에 따른 열값 바꾸기
# - 시간당 강수량의 평균을 구하고 그것을 기준으로 DIST값 지정
# 시간당강우량 0을 제외한 평균값은 10

F_Final_DF[F_Final_DF['HOUR_RAINFALL']>0].mean() 
```

```python
# 조건에 따른 열값 바꾸기
# - 시간당 강수량이 10초과이면 DIST 값이 1

F_Final_DF.loc[(F_Final_DF['HOUR_RAINFALL'] > 10),'DIST'] = 1
```

```python
# 조건에 따른 열값 바꾸기
# - 시간당 강수량이 10이하이면 DIST 값이 0

F_Final_DF.loc[(F_Final_DF['HOUR_RAINFALL'] <= 10),'DIST'] = 0
```

```python
F_Final_DF.to_excel("Reflux_data/Final_V1/F_Final_V2_test2.xlsx")
```
