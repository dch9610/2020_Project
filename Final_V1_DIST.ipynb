## 강수량이 있으면 DIST값이 1인 조건

```python
F_Final_V1 = pd.read_excel('Reflux_data/Final_V1/F_Final_V1.xlsx',index_col=False)
F_Final_V1.head()
```

```python
# 조건에 따른 열값 바꾸기
# - 시간당 강수량이 0이상이면 DIST 값이 1

F_Final_V1.loc[(F_Final_V1['HOUR_RAINFALL'] > 0),'DIST'] = 1
```

```python
# 조건에 따른 열값 바꾸기
# - 시간당 강수량이 0이면 DIST 값이 0

F_Final_V1.loc[(F_Final_V1['HOUR_RAINFALL'] == 0),'DIST'] = 0
```

```python
F_Final_V1
```

```python
F_Final_V1.to_excel("Reflux_data/Final_V1/F_Final_V1_01.xlsx")
```

## 강우량 평균을 기준으로 다른 조건

```python
F_Final_V1 = pd.read_excel('Reflux_data/Final_V1/F_Final_V1.xlsx',index_col=False)
F_Final_V1.head()
```

```python
# 조건에 따른 열값 바꾸기
# - 시간당 강수량의 평균을 구하고 그것을 기준으로 DIST값 지정


F_Final_V1[F_Final_V1['HOUR_RAINFALL']>0].mean() # 1.031633e+01
```

```python
# 조건에 따른 열값 바꾸기
# - 시간당 강수량이 10초과이면 DIST 값이 1

F_Final_V1.loc[(F_Final_V1['HOUR_RAINFALL'] >= 10),'DIST'] = 1
```

```python
# 조건에 따른 열값 바꾸기
# - 시간당 강수량이 10이하이면 DIST 값이 0

F_Final_V1.loc[(F_Final_V1['HOUR_RAINFALL'] < 10),'DIST'] = 0
```

```python
F_Final_V1
```

```python
F_Final_V1.to_excel("Reflux_data/Final_V1/F_Final_V1_02.xlsx")
```
