```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 강우량 > 0인 값만 DIST =1인 데이터

```python
F_Final_DF1 = pd.read_excel('Reflux_data/Final_V1/F_Final_V2_test1.xlsx',index_col=False)
F_Final_DF1
```

```python
Unf_Final_DF1 = pd.read_excel('Reflux_data/Final_V1/Unf_Final_V2_test1.xlsx',index_col=False)
Unf_Final_DF1
```

```python
Final_V2_test1=pd.concat([F_Final_DF1,Unf_Final_DF1])
Final_V2_test1
```

### 상관관계 분석

```python
plt.figure(figsize=(10,10))
sns.heatmap(data = Final_V2_test1.corr(), annot=True, 
fmt = '.2f', linewidths=.5, cmap='Blues')
plt.show()
```

```python
Final_V2_test1.to_excel("Reflux_data/Final_V1/Final_V2_test1.xlsx")
```
