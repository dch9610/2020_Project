# 잘 안되서 다른방식으로 해봄

```python
Test1 = pd.read_excel("Final_V4.xlsx")
```

```python
x1 = Test1.drop(['ZONE','Dong','SEQ','DISTRICT','DIST','Unnamed: 0','ADM_CD','F_BEGIN_DE','F_END_DE','F_BEGIN_TM','F_END_TM','FLUD_TIME','PRECI_HH'], axis= 1)
x2 = Test1['DIST']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x1, x2, test_size=0.3, random_state=100)
print(len(X_train), len(X_test),len(y_train),len(y_test))
```

# 정규화 안한거랑 한거랑 둘다 해볼거

## 아래 과정은 다시 ndarray로 변환해서 deep learning 모델에 사용되는 형태로 바꿔줌

```python

X_train = X_train.values
X_test= X_test.values

y_train = y_train.values
y_test = y_test.values
print(X_train.shape)
print(X_train)
print(y_train.shape)
print(y_train)

# X_train.shape를 프린트해줘서 형태를 확인가능
#나중에 데이터 형식 바꿀때 쓰는거
```

   ## keras 는 RNN 계열 모델을 트레이닝 하기위해  데이터 형식( 3차원)을 요구함

```python
X_train_t = X_train.reshape(X_train.shape[0], 10, 1)
X_test_t = X_test.reshape(X_test.shape[0], 10, 1)

print(X_train_t.shape)
# print(X_train_t)
print(y_train.shape)
```

```python
from keras.layers import LSTM 
from keras.models import Sequential
from keras.layers import Dense
import keras.backend as K 
from keras.callbacks import EarlyStopping

```

```python
K.clear_session() 
model = Sequential() 
model.add(LSTM(20, input_shape=(10, 1))) 
model.add(Dense(1)) 
model.compile(loss='mean_squared_error', optimizer='adam') 

# 일반적인 optimizer 는 아담임
# 예측하는 타겟의 갯수는 DIST 1개니까 DENSE는 1로줌
```

```python
model.summary()
```

```python
early_stop = EarlyStopping(monitor='loss', patience=1, verbose=1)
# 이 객체를 이용해 poch 마다 early stopping 체크한다.
# early stopping 는 무조건 epoch 를 마구 돌린후 특정시점에서 멈추게하는거
```

```python
model.fit(X_train_t, y_train, epochs=100, batch_size=30,
          verbose=1, callbacks=[early_stop])
```

```python
print(X_test_t)
```

```python
y_pred = model.predict(X_test_t)
print(y_pred)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

conf_matrix = confusion_matrix(y_test, y_pred)
print(conf_matrix)

report = classification_report(y_test, y_pred)
print(report)
```

```python
print(model.score(X_train, y_train)) #

print(model.score(X_test, y_test))
```
